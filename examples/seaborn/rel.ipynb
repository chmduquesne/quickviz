First plot
==========

Let's learn quickviz while attempting here to reproduce the graphs of [the corresponding seaborn tutorial](https://seaborn.pydata.org/tutorial/relational.html).

First, we need to import stuff as recommended by seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

Let's also import quickviz

```python
# The magic
from quickviz import visualize as v
```

It has only one function, `visualize`. We recommend to import it as `v`.

Let's reproduce the first visualisation of the seaborn tutorial. Here is how they do it:

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips)
```

Let's try with quickviz: we use `v` on the dataframe:

```python
v(tips)
```

Let's explain what the buttons do. The dropdown described as `Plot type` lets you select the kind of graph you want. Let's go ahead and select `relplot`. The second dropdown, `Add arg` adds arguments to the selected plot type. We select `x`, then `y`. This generates 2 new dropdown lists for `x` and `y`. We select to have `x` as `total_bill` and `y` as `tip`. Last, we check the `auto-update` box.

save/load
---------
This is all very nice, but is we re-run all cells of the notebook, lose the state of the plot. It's time to learn about the `save` feature. Instead of just running `v(tips)`, we can assign the result:

```python
plot = v(tips)
```

I won't bother setting all the parameters again. The point is to show that the variable `plot` has a `save` method:

```python
plot.save()
```

It can also `load` a dictionary that was generated with `save`

```python
plot = v(tips)
plot.load({'plot': 'relplot', 'kwargs': {'x': 'total_bill', 'y': 'tip'}})
```

```python
plot.save()
```

Bam, reproducible results!

Second plot
===========

How seaborn does it:

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

Now that you know about `load`/`save`, I won't spend time to explain how to obtain the same result through clicking and I will directly load the right dictionary

```python
p = v(tips)
p.load({'plot': 'relplot', 'kwargs': {'x': 'total_bill', 'y': 'tip', 'hue': 'smoker'}})
```

Third plot
==========

```python
p = v(tips)
p.load({'plot': 'relplot',
 'kwargs': {'x': 'total_bill', 'y': 'tip', 'hue': 'smoker', 'style': 'smoker'}})
```

Fourth plot
===========

```python
p = v(tips)
p.load({'plot': 'relplot',
 'kwargs': {'x': 'total_bill', 'y': 'tip', 'hue': 'smoker', 'style': 'time'}})
```

Fifth plot
==========

```python
p = v(tips)
p.load({'plot': 'relplot',
 'kwargs': {'x': 'total_bill', 'y': 'tip', 'hue': 'size'}})
```

Sixth plot
==========

```python
p = v(tips)
p.load({'plot': 'relplot',
 'kwargs': {'x': 'total_bill',
  'y': 'tip',
  'hue': 'size',
  'palette': 'ch:r=-.5,l=.75'}})
```
