Quickviz Tutorial
=================

Let's get ourselves a dataframe

```python
from vega_datasets import data
df = data.cars()
df.head()
```

Now let's play! Explanation of the UI:

* The `Plot` dropdown lets you select the kind of graph you want.
* The `Add arg` dropdown, adds arguments to the selected plot type. It creates new widgets.
* These new widgets are named after the argument they control. They control what is assigned to this argument.
* The `auto-update` checkbox lets you see live updates to the plot.
* `auto-update` is unchecked by default to avoid seeing errors when the current plot has not a valid arguments list.

```python
from quickviz import visualize as v
plot = v(df)
```

Saving and Loading the UI state
-------------------------------
You can obtain the state of the UI using the save() method

```python
plot.save()
```

You can load a state into an existing UI with load()

```python
relplot = v(df)
relplot.load({'kwargs': {'hue': 'Origin',
  'size': 'Weight_in_lbs',
  'x': 'Horsepower',
  'y': 'Miles_per_Gallon'},
 'plot': 'relplot'})
```

Arguments that the UI does not handle
-------------------------------------

Some arguments are not handled by the UI. You can pass them to the plot using the `manual_kwargs` dictionary.

```python
import seaborn
df = seaborn.load_dataset("tips")

catplot = v(df)
# We pass inner=None manually
catplot.manual_kwargs = {"inner": None}
catplot.load({'plot': 'catplot',
 'kwargs': {'x': 'day', 'y': 'total_bill', 'kind': 'violin'}})
```
